```python
emoji_wordvector={}
```

```python
f=open("./test data/glove.6B.50d.txt",encoding="utf-8")
```

```python
for line in f:
    lostr=line.split()
    word=lostr[0]
    wordvector=np.asarray(lostr[1:]).astype("float")
    emoji_wordvector[word]=wordvector
    
f.close()    
```

```python
embedding_size=50
emoji_wordvector["this"]
```

```python
# convert 2d matrix into 3d matrix 
maxlength=10
def embedding_layer_output(X):
    
    out=np.zeros((X.shape[0],maxlength,embedding_size))
    for sentno in range(X.shape[0]):
        for ij in range(len(X[sentno])):
            out[sentno][ij] = emoji_wordvector[X[sentno][ij].lower()]
            
        
    return out    
```

```python
temp=embedding_layer_output(Xtrain)
temp.shape
```

```python
from keras.callbacks import ModelCheckpoint
checkpoint=ModelCheckpoint("goodmodel.h5",monitor="val_acc",save_best_only=True)
```

```python
from keras.models import Sequential
from keras.layers import *
from keras.utils import to_categorical
```

```python
model=Sequential()
model.add(LSTM(32,input_shape=(10,50),return_sequences=True))
model.add(Dropout(0.5))
# model.add(LSTM(32,return_sequences=True))
# model.add(Dropout(0.5))
model.add(LSTM(32,return_sequences=False))
model.add(Dropout(0.5))
model.add(Dense(5,activation="softmax"))



```

```python
model.compile(optimizer="rmsprop",loss="categorical_crossentropy",metrics=["acc"])
```

```python
hist=model.fit(embedding_layer_output(Xtrain),Ytrain,epochs=100,batch_size=32,validation_split=0.2,callbacks=[checkpoint])
```
